```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming `volumes` is a list of 3D numpy arrays, each representing an MRI volume
volumes = [np.random.rand(64, 64, 64) for _ in range(4)]  # Example data

def plot_volume_slices(volume, num_slices=16):
    fig, axes = plt.subplots(1, num_slices, figsize=(20, 2))
    slice_interval = volume.shape[2] // num_slices
    
    for i in range(num_slices):
        ax = axes[i]
        slice_index = i * slice_interval
        ax.imshow(volume[:, :, slice_index], cmap='gray')
        ax.axis('off')
        
    plt.tight_layout()
    plt.show()

# Visualize each MRI volume with a reasonable number of slices
for i, volume in enumerate(volumes):
    print(f"Visualization of MRI volume {i+1}:")
    plot_volume_slices(volume)
```

```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from glob import glob
import os

main_directory = "archive/BraTS2021_Training_Data/test/BraTS2021_01559"

# Get a list of all MRI files in the dataset folder
mri_files = glob(os.path.join(main_directory, "*.nii.gz"))

# Define the target size
target_size = (256, 256, 256)

# Initialize lists to store original and resized MRI data
original_mri_data = []
resized_mri_data = []

# Iterate over each MRI file, load its data, and resize it
for file in mri_files:
    # Load the MRI data
    mri_data = nib.load(file).get_fdata()
    
    # Append the original MRI data to the list
    original_mri_data.append(mri_data)
    
    # Get the shape of the original MRI data
    original_size = mri_data.shape
    
    # Create an array of zeros with the target size
    padded_mri_data = np.zeros(target_size)
    
    # Calculate the starting indices to copy the original MRI data into the padded array
    start_indices = tuple((np.array(target_size) - np.array(original_size)) // 2)
    
    # Calculate the ending indices to copy the original MRI data into the padded array
    end_indices = tuple(start_indices + np.array(original_size))
    
    # Copy the original MRI data into the center of the padded array
    padded_mri_data[start_indices[0]:end_indices[0], 
                    start_indices[1]:end_indices[1], 
                    start_indices[2]:end_indices[2]] = mri_data
    
    # Append the padded MRI data to the list
    resized_mri_data.append(padded_mri_data)

# Convert the original and padded MRIs to numpy arrays
original_mri_data = np.array(original_mri_data)
resized_mri_data = np.array(resized_mri_data)

print("Original MRIs shape:", [data.shape for data in original_mri_data])
print("Resized MRIs shape:", resized_mri_data.shape)

# Define a function to plot a single slice
def plot_slices(original_volume, resized_volume, slice_index, title):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    axes[0].imshow(original_volume[slice_index,:, :], cmap='gray')
    axes[0].set_title('Original ' + title)
    axes[0].axis('off')
    
    axes[1].imshow(resized_volume[:, slice_index, :], cmap='gray')
    axes[1].set_title('Resized ' + title)
    axes[1].axis('off')
    
    plt.show()

# Calculate the middle slice index for the second axis
middle_slice_index = target_size[1] // 2

# Plot the middle slice of each original and resized volume
for i, (original_volume, resized_volume) in enumerate(zip(original_mri_data, resized_mri_data)):
    plot_slices(original_volume, resized_volume, middle_slice_index, f'Volume {i+1} - Middle Slice')
```
